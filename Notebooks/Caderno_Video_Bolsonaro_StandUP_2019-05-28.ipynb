```python
#importar bibliotecas 

import pandas as pd
from matplotlib import pyplot as plt
from datetime import datetime
```

```python
#configurar o api

api_key = "[inserir o seu api_Key]"

from googleapiclient.discovery import build
youtube = build("youtube", "v3", developerKey=api_key)
```

```python
q_vid = youtube.commentThreads().list(videoId ="AkKKJfuFQBQ", part = "snippet",  order = "relevance", maxResults = 100)
comment_FLV = q_vid.execute()
```

```python
comment_FLV['items'][0]
```

```python
#Explorar

for r in comment_FLV['items']:
    print(r['snippet']['topLevelComment']['snippet']['publishedAt'],"-",
          r['snippet']['topLevelComment']['snippet']['authorDisplayName'],
          r['snippet']['topLevelComment']['snippet']['authorChannelUrl'],
          r['snippet']['topLevelComment']['snippet']['likeCount'], "-L/",
          r['snippet']['totalReplyCount'],  "-R/" , "\n",
          r['snippet']['topLevelComment']['snippet']['textOriginal'], 
          "\n")
```

```python
#Salvar

Coment_FLV_Sup = []

for r in comment_FLV['items']:
    PublishedAt = r['snippet']['topLevelComment']['snippet']['publishedAt']
    Name = r['snippet']['topLevelComment']['snippet']['authorDisplayName']
    AuthorChannelUrl = r['snippet']['topLevelComment']['snippet']['authorChannelUrl']
    Like = r['snippet']['topLevelComment']['snippet']['likeCount']
    Reply = r['snippet']['totalReplyCount']
    Text = r['snippet']['topLevelComment']['snippet']['textOriginal']
    Coment_FLV_Sup.append((PublishedAt, Name, AuthorChannelUrl, Like, Reply, Text))
    
Coment_FLV_Sup = pd.DataFrame(Coment_FLV_Sup, columns= ["PublishedAt", "Name", "AuthorChannelUrl", "Like",  "Reply", "Text"])
```

```python
Coment_FLV_Sup["PublishedAt"] = pd.to_datetime(Coment_FLV_Sup["PublishedAt"])
```
