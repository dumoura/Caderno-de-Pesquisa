```python
# criar fun√ß√£o para pegar todos os v√≠deos de uma canal
# 1. fazer o request para o channel id
# 2. get o id dos playlistId
# 3. fazer um looping para pegar todos os v√≠deos do canal (para isso, criar um treco vazio chamado v√≠deo e algo que passe para o looping para a pr√≥xima p√°gina do yt)
```

```python
def get_channel_videos(channel_id):
    
    # get Uploads playlist id
    res = youtube.channels().list(id=channel_id, 
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id, 
                                           part='snippet', 
                                           maxResults=50,
                                           pageToken=next_page_token).execute()
       
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
    
    return videos
```

```python
#esse e o link do tutorial, com o bolsonaro n√£o deu certo :/ 
videos = get_channel_videos('UCQziFSWcDRTVQVnsrDa7v0A')
```

```python
video_ids = list(map(lambda x:x ['snippet']['resourceId']['videoId'], videos))
```

```python
def get_vid_stats (video_ids):
    stats = []
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(id="," .join(video_ids[i:i+50]),
                                   part='statistics').execute()
        
        stats += res['items']
    
    return stats
```

```python
stats = get_vid_stats(video_ids)
```

```python
stats[0]
```

```python
most_disliked = sorted(stats, key=lambda x:int(x['statistics']['dislikeCount']), reverse=True)
```

```python
most_disliked
```

```python
for i in most_disliked:
    print(i['statistics']['dislikeCount'], i['statistics']['likeCount'], "https://www.youtube.com/watch?v="+i['id'])
```

```python
most_liked = sorted(stats, key=lambda x:int(x['statistics']['likeCount']), reverse=True)
```

```python
for i in most_liked :
    print(i['statistics']['dislikeCount'], i['statistics']['likeCount'], "https://www.youtube.com/watch?v="+i['id'])
```

```python
df_likeCount = []
for i in most_liked:
    dislike = i['statistics']['dislikeCount']
    like = i['statistics']['likeCount']
    link = ("https://www.youtube.com/watch?v="+i['id'])
    df_likeCount.append((dislike, like, link))
    

dfVC = pd.DataFrame(df_likeCount, columns=['dislike', 'like' ,'link'])
dfVC.head()
```
